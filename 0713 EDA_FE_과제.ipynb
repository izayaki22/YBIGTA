### 5) Feature Engineering

feature에 대한 전처리를 하기 이전에 결측치가 있는 column을 모두 제거했는데, 다시 초기 상태로 되돌리고 진행하겠습니다.

```python
df = pd.read_csv('drive/MyDrive/airbnb_listings.csv', index_col=0)
```

- #### <font color='orange'> Q12. 결측치에 대한 처리를 하고 싶습니다.
#### <font color='orange'> 주어진 모든 column들에 대하여 결측치가 25%이상인 경우에는 column을 drop하고 10%이상 25%미만인 경우에는 결측치가 있는 row만 drop하고, 10%미만인 경우에는 중앙값으로 대체해주세요.

```python
#Q12

row_drop=[]

for col in df.columns:
   miss_rate = df[col].isnull().sum() / len(df[col])
   if(miss_rate >= 0.25):
    df.drop(col, axis=1, inplace=True)

   elif(miss_rate >= 0.10):
    row_drop.append(col)

   elif(miss_rate < 0.10 and df[col].dtypes != object):
    median = df[col].median()
    df[col].fillna(median, inplace=True)

df.dropna(subset=row_drop, inplace=True)
```

- #### <font color='orange'> Q13. 앞서 진행한 EDA를 바탕으로 FE를 진행하고자 합니다..
#### <font color='orange'> 총 3개의 FE를 진행하시고, 해당 FE를 진행한 이유에 대해 **간단히** 작성해주세요.


```python
#Q13-1) 문자열을 숫자열로 바꾸어주는 Label Encoding을 진행하였다. 이전에는 object이기 때문에 결측치를 중간 값 등으로 대체할 수 없었기 때문이다.

is_superhost = []
for s in df['host_is_superhost']:
    if (s == 'f'):
        is_superhost.append(0)
    else:
        is_superhost.append(1)

df['host_is_superhost'] = is_superhost

display(df)
```

```python
#Q13-2) review_scores_rating을 very good, good, so so, bad로 구분하여 지표 변수를 만들었다. 점수에 따라 각 숙소가 얼마나 좋은지 한눈에 바로 알 수 있기 때문이다.

evaluation = []
for r in df['review_scores_rating']:
    if (r >= 95):
        evaluation.append('very good')
    elif (r >= 80):
        evaluation.append('good')
    elif (r >= 60):
        evaluation.append('so so')
    else:
        evaluation.append('bad')

df['evaluation'] = evaluation

display(df)
```

```python
#Q13-3) evalution 열에 one-hot encoding을 진행하였다. 범주형 자료를 정보 손실 없이 그룹화할 수 있기 때문이다.

pd.get_dummies(df['evaluation'])
```
